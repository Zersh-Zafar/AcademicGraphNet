```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Sample labeled data (simulate OS textbook phrases)
train_data = [
    ("Linux uses paging for process management",
     ["OPERATING_SYSTEM", "O", "OS_COMPONENT/ALGORITHM", "O", "OS_CONCEPT", "OS_CONCEPT"]),
    ("Windows manages files and memory efficiently",
     ["OPERATING_SYSTEM", "O", "OS_COMPONENT/ALGORITHM", "O", "OS_CONCEPT", "O"]),
    ("CPU and RAM are essential hardware devices",
     ["HARDWARE/DEVICE", "O", "HARDWARE/DEVICE", "O", "O", "HARDWARE/DEVICE", "HARDWARE/DEVICE"]),
    ("UNIX provides multitasking and multiuser capabilities",
     ["OPERATING_SYSTEM", "O", "OS_CONCEPT", "O", "OS_CONCEPT", "O"]),
]
```

```python
train_words = []
train_labels = []

for sent, labels in train_data:
    words = sent.split()
    for w, l in zip(words, labels):
        train_words.append(w.lower())
        train_labels.append(l)

le_words = LabelEncoder()
le_labels = LabelEncoder()

word_ids = le_words.fit_transform(train_words)
label_ids = le_labels.fit_transform(train_labels)

num_states = len(le_labels.classes_)
num_obs = len(le_words.classes_)

print(f"✅ Vocabulary size: {num_obs}, Labels: {num_states}")
```

```python
# Initialize count matrices
trans_counts = np.zeros((num_states, num_states))
emit_counts = np.zeros((num_states, num_obs))
start_counts = np.zeros(num_states)

# Count transitions and emissions
for i in range(len(label_ids)-1):
    trans_counts[label_ids[i], label_ids[i+1]] += 1
    emit_counts[label_ids[i], word_ids[i]] += 1
start_counts[label_ids[0]] += 1

# Convert to probabilities with smoothing
trans_prob = (trans_counts + 1) / (trans_counts.sum(axis=1, keepdims=True) + num_states)
emit_prob = (emit_counts + 1) / (emit_counts.sum(axis=1, keepdims=True) + num_obs)
start_prob = (start_counts + 1) / (start_counts.sum() + num_states)

print("✅ HMM probabilities estimated successfully.")
```

```python
def viterbi_fast(sentence):
    words = sentence.lower().split()
    obs = []
    for w in words:
        if w in le_words.classes_:
            obs.append(le_words.transform([w])[0])
        else:
            obs.append(0)  # unknown word fallback

    T = len(obs)
    delta = np.zeros((T, num_states))
    psi = np.zeros((T, num_states), dtype=int)

    delta[0] = np.log(start_prob) + np.log(emit_prob[:, obs[0]])

    for t in range(1, T):
        prev = delta[t-1].reshape(-1, 1) + np.log(trans_prob)
        psi[t] = np.argmax(prev, axis=0)
        delta[t] = np.max(prev, axis=0) + np.log(emit_prob[:, obs[t]])

    states = np.zeros(T, dtype=int)
    states[-1] = np.argmax(delta[-1])
    for t in reversed(range(1, T)):
        states[t-1] = psi[t, states[t]]

    return [le_labels.classes_[s] for s in states]
```

```python
test_sentence = "Linux uses threads for process management"
decoded_labels = viterbi_fast(test_sentence)

print("\n📌 HMM Prediction:")
for w, l in zip(test_sentence.split(), decoded_labels):
    print(f"{w:15} → {l}")
```

```python
def word2features(sent, i):
    word = sent[i]
    features = {
        'word.lower()': word.lower(),
        'word.isupper()': word.isupper(),
        'word.istitle()': word.istitle(),
        'word.isdigit()': word.isdigit(),
    }
    if i > 0:
        features.update({
            '-1:word.lower()': sent[i-1].lower(),
            '-1:word.istitle()': sent[i-1].istitle(),
        })
    else:
        features['BOS'] = True  # Beginning of sentence
    if i < len(sent)-1:
        features.update({
            '+1:word.lower()': sent[i+1].lower(),
            '+1:word.istitle()': sent[i+1].istitle(),
        })
    else:
        features['EOS'] = True  # End of sentence
    return features
```

```python
# ============================================
# ✅ FIXED MEMM Implementation (No Skipping)
# ============================================

import nltk
from nltk import word_tokenize, pos_tag
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# -----------------------------
# 🔧 Step 1: Download NLTK Data
# -----------------------------
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')

# -----------------------------
# 🧠 Step 2: Corrected Training Data
# -----------------------------
train_sentences = [
    ("Windows uses semaphores for process synchronization .",
     ["OPERATING_SYSTEM", "O", "OS_COMPONENT/ALGORITHM", "O", "PROGRAMMING_CONCEPT", "PROGRAMMING_CONCEPT", "O"]),
    ("Linux supports paging and deadlock handling .",
     ["OPERATING_SYSTEM", "O", "OS_COMPONENT/ALGORITHM", "O", "OS_COMPONENT/ALGORITHM", "PROGRAMMING_CONCEPT", "O"]),
    ("Dijkstra developed the Banker 's algorithm .",
     ["PERSON", "O", "O", "O", "O", "OS_COMPONENT/ALGORITHM", "O"]),
    ("Semaphore is used for process synchronization .",
     ["OS_COMPONENT/ALGORITHM", "O", "O", "O", "PROGRAMMING_CONCEPT", "PROGRAMMING_CONCEPT", "O"]),
    ("CPU scheduling improves system performance .",
     ["HARDWARE/DEVICE", "OS_COMPONENT/ALGORITHM", "O", "O", "PROGRAMMING_CONCEPT", "O"]),
    ("Charles Babbage invented computer .",
     ["PERSON", "PERSON", "O", "HARDWARE/DEVICE", "O"]),
    ("Windows handles deadlock using resource allocation .",
     ["OPERATING_SYSTEM", "O", "OS_COMPONENT/ALGORITHM", "O", "PROGRAMMING_CONCEPT", "PROGRAMMING_CONCEPT", "O"]),
    ("Linux manages memory and I/O operations .",
     ["OPERATING_SYSTEM", "O", "OS_COMPONENT/ALGORITHM", "O", "OS_COMPONENT/ALGORITHM", "PROGRAMMING_CONCEPT", "O"]),
]

# -----------------------------
# ⚙️ Step 3: Feature Extraction
# -----------------------------
def word2features(tokens, i):
    word, pos = tokens[i]
    features = {
        'word.lower()': word.lower(),
        'is_first': i == 0,
        'is_last': i == len(tokens) - 1,
        'is_capitalized': word[0].upper() == word[0],
        'pos_tag': pos,
        'prev_word': '' if i == 0 else tokens[i - 1][0].lower(),
        'next_word': '' if i == len(tokens) - 1 else tokens[i + 1][0].lower(),
    }
    return features

# -----------------------------
# 🧩 Step 4: Prepare Data
# -----------------------------
X_train, y_train = [], []
for sentence, labels in train_sentences:
    tokens = pos_tag(word_tokenize(sentence))
    for i in range(len(tokens)):
        X_train.append(word2features(tokens, i))
        y_train.append(labels[i])

# -----------------------------
# 🚀 Step 5: Train MEMM Model
# -----------------------------
vec = DictVectorizer(sparse=True)
X_train_vec = vec.fit_transform(X_train)

memm = LogisticRegression(max_iter=300)
memm.fit(X_train_vec, y_train)

print("✅ MEMM training complete.\n")

# -----------------------------
# 📊 Step 6: Evaluation
# -----------------------------
y_train_pred = memm.predict(X_train_vec)
print("📊 Evaluation Report (on train data):")
print(classification_report(y_train, y_train_pred))
print(f"🎯 Overall Accuracy: {accuracy_score(y_train, y_train_pred) * 100:.2f}%\n")

# -----------------------------
# 🧪 Step 7: Test on New Sentence
# -----------------------------
test_sentence = "Linux uses semaphore and paging"
tokens = pos_tag(word_tokenize(test_sentence))
X_test = [word2features(tokens, i) for i in range(len(tokens))]
X_test_vec = vec.transform(X_test)
y_test_pred = memm.predict(X_test_vec)

print("📌 MEMM Prediction on New Sentence:")
for w, l in zip([t[0] for t in tokens], y_test_pred):
    print(f"{w:15} → {l}")
```

```python
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^A-Za-z0-9.,!? ]+', '', text)
    return text.lower()

cleaned_text = clean_text(all_text)
```

```python
train_data, test_data = train_test_split(
    labeled_data, test_size=0.2, random_state=42
)

print("Train sentences:", len(train_data))
print("Test sentences:", len(test_data))
```

```python
class HMM:
    def __init__(self):
        self.emission = defaultdict(Counter)
        self.tag_counts = Counter()

    def train(self, data):
        for sent in data:
            for word, tag in sent:
                self.emission[tag][word] += 1
                self.tag_counts[tag] += 1

    def predict(self, sentence):
        predictions = []
        for word in sentence:
            best_tag = max(
                self.tag_counts,
                key=lambda t: self.emission[t][word]
            )
            predictions.append(best_tag)
        return predictions

hmm = HMM()
hmm.train(train_data)
```

```python
y_true = []
y_pred = []

for sent in test_data:
    words = [w for w, t in sent]
    true_tags = [t for w, t in sent]
    pred_tags = hmm.predict(words)

    y_true.extend(true_tags)
    y_pred.extend(pred_tags)

print("HMM Accuracy:", accuracy_score(y_true, y_pred))
```

```python
# OS domain keyword dictionaries

topic_words = [
    "process", "thread", "deadlock", "memory", "filesystem",
    "scheduling", "virtualization", "security"
]

concept_words = [
    "critical section", "context switch", "race condition",
    "paging", "segmentation", "mutual exclusion"
]

method_words = [
    "system call", "interrupt", "fork", "exec", "wait"
]

algorithm_words = [
    "banker", "fcfs", "sjf", "round robin",
    "priority scheduling", "lru"
]
```

```python
def label_token(word):
    w = word.lower()

    if w in topic_words:
        return "topic"
    elif w in concept_words:
        return "concept"
    elif w in method_words:
        return "method"
    elif w in algorithm_words:
        return "algorithm"
    else:
        return "other"
```

```python
labeled_data = []

for sent in tokenized:
    labeled_sentence = []
    for word in sent:
        tag = label_token(word)
        labeled_sentence.append((word, tag))
    labeled_data.append(labeled_sentence)

print("Example labeled sentence:")
print(labeled_data[0][:10])
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(
    labeled_data, test_size=0.2, random_state=42
)

print("Train sentences:", len(train_data))
print("Test sentences:", len(test_data))
```

```python
from collections import defaultdict, Counter

class HMM:
    def __init__(self):
        self.emission = defaultdict(Counter)
        self.transition = defaultdict(Counter)
        self.tag_counts = Counter()

    def train(self, data):
        for sent in data:
            prev_tag = "<s>"
            for word, tag in sent:
                self.transition[prev_tag][tag] += 1
                self.emission[tag][word] += 1
                self.tag_counts[tag] += 1
                prev_tag = tag
            self.transition[prev_tag]["</s>"] += 1

    def predict(self, sentence):
        predictions = []
        for word in sentence:
            best_tag = max(
                self.tag_counts,
                key=lambda t: self.emission[t][word]
            )
            predictions.append(best_tag)
        return predictions

hmm = HMM()
hmm.train(train_data)
```

```python
from sklearn.metrics import accuracy_score

y_true = []
y_pred = []

for sent in test_data:
    words = [w for w, t in sent]
    true_tags = [t for w, t in sent]
    pred_tags = hmm.predict(words)

    y_true.extend(true_tags)
    y_pred.extend(pred_tags)

print("HMM OS Label Accuracy:", accuracy_score(y_true, y_pred))
```

```python
from sklearn.metrics import classification_report

# Generate a detailed report
report = classification_report(y_true, y_pred, labels=["topic", "concept", "method", "algorithm", "other"])
print("HMM OS Label Classification Report:\n")
print(report)
```
