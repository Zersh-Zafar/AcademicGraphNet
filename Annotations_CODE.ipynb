```python
import pdfplumber

pages = []

with pdfplumber.open("/content/Network Devices.pdf") as pdf:
    print("Total PDF pages:", len(pdf.pages))

    for page in pdf.pages:
        text = page.extract_text()
        if text:
            pages.append(text.strip())

print("Pages extracted:", len(pages))
```

```python
# ==============================
# COMPUTER NETWORK MASTER DICTIONARY (CORRECTED)
# ==============================

TOPIC = [

"computer networks","network architecture","network models",
"osi model","tcp ip model","physical layer","data link layer",
"network layer","transport layer","session layer",
"presentation layer","application layer",

"network routing","network switching",
"wireless networking","optical networking","mobile networking",

"network security","network management",
"cloud networking","data center networking","internet architecture",
"internet of things networking","software defined networking",
"network virtualization","edge networking",
"multimedia networking","peer to peer networking",
"distributed networking","ad hoc networking",

"network performance","traffic engineering"
]


CONCEPT = [

# Core Communication Concepts
"protocol","packet","frame","datagram","segment",

# Network Topology
"network topology","bus topology","star topology",
"ring topology","mesh topology","tree topology",
"hybrid topology",

# Addressing
"logical address","physical address","ip address",
"mac address","subnet mask","default gateway",

# Data Transmission
"bandwidth","throughput","latency","jitter",
"packet loss","propagation delay","transmission delay",

# Data Handling
"encapsulation","decapsulation","fragmentation",
"reassembly","multiplexing","demultiplexing",

# Traffic & Flow
"congestion","traffic load","network delay",
"quality of service",

# Communication Types
"unicast communication","multicast communication",
"broadcast communication","anycast communication",

# Reliability
"error detection","error control","checksum",
"cyclic redundancy check","parity bit",

# System Properties
"scalability","availability","reliability",
"redundancy",

# System Design Concepts
"collision detection","collision avoidance",
"network reliability","fault tolerance","high availability"
]


METHOD = [

# Switching Techniques
"packet switching","circuit switching","message switching",

# Multiplexing Methods
"time division multiplexing","frequency division multiplexing",
"wavelength division multiplexing","statistical multiplexing",

# Flow Control Methods
"stop and wait","sliding window",

# ARQ Methods
"go back n arq","selective repeat arq",

# Medium Access Techniques
"carrier sense multiple access","token passing","polling",

# Network Management Techniques
"traffic shaping","traffic policing","load balancing",
"packet filtering",

# Security Techniques
"encryption","authentication",
"firewall filtering","stateful inspection",
"intrusion detection",

# Routing Methods
"distance vector routing","link state routing",
"path vector routing",

# System Methods
"load distribution","network monitoring"
]


ALGORITHM = [

# Routing Algorithms
"dijkstra algorithm","bellman ford algorithm",
"distance vector algorithm","link state algorithm",
"shortest path first algorithm",

# Ethernet Algorithms
"csma cd algorithm","csma ca algorithm",

# Traffic Control
"leaky bucket algorithm","token bucket algorithm",
"random early detection algorithm",

# Queue Scheduling
"fair queuing algorithm","weighted fair queuing algorithm",
"round robin algorithm",

# Network Tree
"spanning tree algorithm",

# Flooding
"flooding algorithm",

# Congestion Control
"backoff algorithm"
]


OTHER = [

# Network Devices
"router","switch","hub","bridge","gateway",
"repeater","modem","access point","firewall",
"proxy server","load balancer","network interface card",

# Internet Protocols
"tcp","udp","ip","ipv4","ipv6",
"http","https","ftp","smtp","pop3","imap",
"dns","dhcp","arp","rarp","icmp","snmp",

# Routing Protocols
"rip","ospf","bgp","isis",

# Wireless Standards
"wifi","bluetooth","zigbee","lte","5g",

# Networking Organizations
"ieee","ietf","iso",

# Transmission Technologies
"ethernet","fiber optics","coaxial cable","twisted pair"
]


label_dict = {
    "TOPIC": TOPIC,
    "CONCEPT": CONCEPT,
    "METHOD": METHOD,
    "ALGORITHM": ALGORITHM,
    "OTHER": OTHER
}
```

```python
import json
import re

# ---------- TEXT CLEANING ----------
def clean_text(text):
    text = text.replace("-\n", "")   # fix hyphen split
    text = text.replace("\n", " ")
    return text

# ---------- ENTITY EXTRACTION ----------
def extract_entities(text):

    text = clean_text(text)
    found_entities = []
    seen = set()

    # Sort longest phrases first (avoid partial match problem)
    for label, keywords in label_dict.items():

        sorted_keywords = sorted(keywords, key=lambda x: len(x), reverse=True)

        for word in sorted_keywords:

            if label == "METHOD":
                # Exact match only for METHOD
                pattern = r'\b' + re.escape(word) + r'\b'
                matches = re.finditer(pattern, text, re.IGNORECASE)

                for match in matches:
                    start = match.start()
                    end = match.end()

                    context_window = text[max(0, start-15):min(len(text), end+15)].lower()

                    # 🔥 Only accept if real system call context
                    if (
                        word.lower() + "(" in text.lower() or
                        "system call" in context_window or
                        match.group().islower()
                    ):
                        key = (match.group().lower(), label)
                        if key not in seen:
                            found_entities.append({
                                "entity": match.group(),
                                "label": label
                            })
                            seen.add(key)

            else:
                # Other labels allow plural s
                pattern = r'\b' + re.escape(word) + r's?\b'
                matches = re.findall(pattern, text, re.IGNORECASE)

                for match in matches:
                    key = (match.lower(), label)
                    if key not in seen:
                        found_entities.append({
                            "entity": match,
                            "label": label
                        })
                        seen.add(key)

    if not found_entities:
        found_entities.append({
            "entity": "No_OS_Entity_Found",
            "label": "OTHER"
        })

    return found_entities


# ---------- CREATE JSONL FILE ----------
annotated_data = []

for page in pages:
    entities = extract_entities(page)

    json_obj = {
        "instruction": "Extract entities using ONLY the following labels: TOPIC, CONCEPT, METHOD, ALGORITHM, OTHER.",
        "input": clean_text(page.strip()),
        "output": entities
    }

    annotated_data.append(json_obj)


# ---------- SAVE FILE ----------
output_file = "output.jsonl"

with open(output_file, "w", encoding="utf-8") as f:
    for item in annotated_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")

print("✅ Annotation Complete")
print("Total documents annotated:", len(annotated_data))
print("Output file saved as:", output_file)
```
